```python
def get_code_len_list(patient_dict):
    
    code_len_list = []
    
    for pid, v in patient_dict.items():
        
        code_list = []
        for visit in v["code_ind"]:
            code_list.extend(visit)
        
        code_len_list.append(len(code_list))
        
    return code_len_list

def count_occurrence_distance(patient_record):
    
    occurrence_distance_list = []
    
    for pid, item in patient_record.items():
        
        occurrence_distance_list.append(item["occurrence_distance"])
        
    return occurrence_distance_list
```

```python
def calculate_stat(rec, var_idx):
    
    return np.sum(rec["code_binary"][:,np.array(var_idx)], axis=0), rec["code_binary"].shape[0]
```

```python
def calculate_stratification_stat(patient_record, occurrence_distance, confounding_var_idx):
    
    var_1_occurrence_count = []
    rec_len_list = []
        
    for pid, item in patient_record.items():
        
        if item["occurrence_distance"] == occurrence_distance:
            confounding_var_occurrence, rec_len = calculate_stat(item, confounding_var_idx)
            var_1_occurrence_count.append(confounding_var_occurrence)
            rec_len_list.append(rec_len)
            
    print("var {} avg occurrence: {}".format(confounding_var_idx, np.mean(var_1_occurrence_count)))
    print("avg record len: {}".format(np.mean(rec_len)))    
```

```python
def get_ps_distribution(patient_record):
    
    treated_ps = []
    control_ps = []
    
    for pid, item in tqdm(patient_record.items()):
                
        if item["treatment_assignment"] == 1:
            treated_ps.append(item["treatment_prob"])
        else:
            control_ps.append(item["treatment_prob"])
            
    return treated_ps, control_ps
```
